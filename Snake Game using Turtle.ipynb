```python
import turtle as t
import random as rd
import time
from playsound import playsound

#screen initialization
screen=t.Screen()
screen.title("Snake Game by Sonali")
screen.bgcolor("yellow")
screen.setup(width=800, height=600)
screen.tracer(0) #turns off animaion of the screen

#snake parameters
snake= t.Turtle()
snake.shape('square')
snake.speed(3) #animation speed
snake.color('red')
snake.penup()
snake.goto(-350,200)
snake.direction = 'stop'  #to initialize snake's direction

#food parameters
food=t.Turtle()
food.speed(0)
#shapes=rd.choice("triangle", 'square','cirle')
#colors=rd.choice("yellow",'green','blue')
food.shape('circle')
food.color('green')
food.penup()

#scores
score=0
high_score=0
pen=t.Turtle()
pen.speed(0)
pen.penup()
pen.color('black')
pen.goto(0,270)
pen.hideturtle()
pen.write("Score : {} High Score : {}".format(score, high_score), align='center', font=('Courier', 20, 'normal'))


segment=[]  #this is for snake body initialization

#Define functions
def snake_down():
    if snake.direction !='Up': #bcoz if we press 'down' and again 'up', the snake reverses direcion. 
        #hence, we check not condition for its opposite direction              
        snake.direction="Down"
    
def snake_up():
    if snake.direction !='Down':
        snake.direction="Up"
    
def snake_left():
    if snake.direction !='Right':
        snake.direction="Left"    
    
def snake_right():
    if snake.direction !='Left':
        snake.direction="Right"

def move():
    
    if snake.direction =='Up':
        y=snake.ycor()
        snake.sety(y + 10)
        
    if snake.direction =='Down':
        y=snake.ycor()
        snake.sety(y - 10)
        
    if snake.direction =='Left':
        x=snake.xcor()
        snake.setx(x - 10)
        
    if snake.direction =='Right':
        x=snake.xcor()
        snake.setx(x + 10)
                
delay=0.1  #0.1 second

#Keyboard entry
screen.listen()
screen.onkeypress(snake_down, 'Down')
screen.onkeypress(snake_up, 'Up')
screen.onkeypress(snake_left, 'Left')
screen.onkeypress(snake_right, '0')

#Main game loop
while True:
    screen.update()
    
    #check for border collisions and bring back to screen from opposite side
    if snake.xcor() > 400 or snake.xcor() < -400 or snake.ycor() > 300 or snake.ycor() < -300:
        if snake.xcor() > 400:
            x=snake.xcor()
            y=snake.ycor()
            snake.goto(-x,y)
        
        elif snake.xcor() < -400:
            x=snake.xcor()
            y=snake.ycor()
            snake.goto(-x,y)    
        
        elif snake.ycor() > 300:
            x=snake.xcor()
            y=snake.ycor()
            snake.goto(x,-y)
        
        elif snake.ycor() < -300:
            x=snake.xcor()
            y=snake.ycor()
            snake.goto(x,-y)
       
    #check food and snake collisions
    if snake.distance(food) < 20: #as each block is of 20 pixels, we calculate distance from snake to food. 
        #it should be less than 20 pixels            
        x=rd.randint(-390, 390) #half of width  -10
        y=rd.randint(-290, 290) #half of height -10
        #move the food to random spot
        food.goto(x,y)
        playsound('bounce.wav')
        
        #create snake body
        new_segment=t.Turtle()
        new_segment.speed(0)
        new_segment.shape('square')
        new_segment.color('grey')
        new_segment.penup()
        segment.append(new_segment)
        
        score += 10
        
        if score> high_score:
            high_score=score
            
        pen.clear()
        pen.write("Score : {} High Score : {}".format(score, high_score), align='center', font=('Courier', 20, 'normal'))
        
        
    #attach the body to snake in reverse order
    
    for index in range(len(segment)-1, 0, -1):
        x = segment[index-1].xcor()
        y = segment[index-1].ycor()
        segment[index].goto(x,y)
        
    #for 1st part of body i.e segment 0
    if len(segment) > 0:
        x=snake.xcor()
        y=snake.ycor()
        segment[0].goto(x,y)
        
    move()
    
    
      
    #self collision code
    #for part in segment:
        #if part.distance(snake)<10:
         #   time.sleep(1)
         #   snake.goto(0,0)
        
          #  for i in segment:
          #      i.goto(1000,1000)
        
       # segment.clear()
            
       
    time.sleep(delay)
    
screen.mainloop()   
```
