# Pong Game

```python
import turtle

from playsound import playsound

#screen initization
wn=turtle.Screen()
wn.title("Pong by Sonali")
wn.bgcolor("purple")
wn.setup(width=800, height=600)
wn.tracer(0) #it gives turle animation a particular delay


# Paddle A
paddle_a= turtle.Turtle()
paddle_a.speed(0) #maximum possible speed
paddle_a.shape("square")
paddle_a.shapesize(stretch_wid=5, stretch_len=1)
paddle_a.color("white")
paddle_a.penup()  #turle will only move on screen, but will not draw anything
paddle_a.goto(-350, 0) #-350 is half of 600(height), cordinates for paddle A

# Paddle B
paddle_b= turtle.Turtle()
paddle_b.speed(0)
paddle_b.shape("square")
paddle_b.shapesize(stretch_wid=5, stretch_len=1)
paddle_b.color("white")
paddle_b.penup()  
paddle_b.goto(350, 0) 

# Ball
ball= turtle.Turtle()
ball.speed(0)
ball.shape("circle")
ball.color('black')
ball.penup() 
ball.goto(0, 0) 
ball.dx=0.2 #the amount by which the turtle's x-cor or y-cor would change
ball.dy=0.2

#scores
score_a=0
score_b=0

#score pen
pen=turtle.Turtle()
pen.speed(0)
pen.penup()
pen.color('white')
pen.goto(0,270)
pen.hideturtle()  #This method is used to make the turtle invisible
pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align ='center', font=('Arial' , 20, 'normal'))




#define functions
def paddle_a_up():
    y=paddle_a.ycor() #take Y-cordinate of paddle A
    y += 20  # increase it by 20 pixel
    paddle_a.sety(y)  #set new Y-cordinate

def paddle_a_down():
    y=paddle_a.ycor() 
    y -= 20 
    paddle_a.sety(y) 

def paddle_b_up():
    y=paddle_b.ycor() 
    y += 20  
    paddle_b.sety(y)  

def paddle_b_down():
    y=paddle_b.ycor() 
    y -= 20 
    paddle_b.sety(y)    

def exitprogram(): #final exit from game
    wn.bye()

def close(): #to ask again for close 
    close = turtle.Turtle()
    close.speed(0)
    #close.color("white")
    close.penup()
    close.hideturtle()
    close.goto(0,0)
    close.write("Press ESC again to exit", align="center", font = ("Courier", 24, "normal"))
    wn.listen()
    wn.onkeypress(exitprogram, "Escape")

#keyboard entry
wn.listen()    #here we make code to listen the keyboard entry
wn.onkeypress(paddle_a_up, 'w')   #we define a key press to particular task
wn.onkeypress(paddle_a_down, 's')  
wn.onkeypress(paddle_b_up, 'Up')   
wn.onkeypress(paddle_b_down, 'Down')
wn.onkeypress(close, "Escape")


#Main game loop
while True:
    wn.update()
    
    #move the ball
    ball.setx(ball.xcor() +ball.dx)
    ball.sety(ball.ycor() +ball.dy)
    
    #Border checking, so that ball does not go out of screen.
    if ball.ycor()>290:
        ball.sety(290)
        ball.dy *= -1
        
        
    if ball.ycor()<-290:
        ball.sety(-290)
        ball.dy *= -1
        
        
        
    if ball.xcor()>390:
        ball.dx *= -1
        score_a += 1
        pen.clear()
        playsound('bounce.wav')
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align ='center', font=('Arial' , 20, 'normal'))
            
    if ball.xcor()<-390:
        ball.dx *= -1
        score_b+=1
        pen.clear()
        playsound('bounce.wav')
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), align ='center', font=('Arial' , 20, 'normal')) 
            
            
    #Paddle and ball collisions
    if (ball.xcor()> 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor() +40 and ball.ycor() > paddle_b.ycor() - 40):
        ball.dx *=-1
        
    if (ball.xcor()< -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor() +40 and ball.ycor() > paddle_a.ycor() - 40):
        ball.dx *=-1    
       
    
        
        
        
        
        
        
        
```
