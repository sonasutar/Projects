```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tkinter import *


sentiment=SentimentIntensityAnalyzer() #create object

root=Tk()
root.title("Sentiment Analysis")
root.geometry('600x500')
root.configure(bg='light green')

#define functions
def detect_sentiment():
    
    sentence=data.get('1.0','end')  #1.0 means that the input should be read from line one, character zero
                                    # end : Read until the end of the text
        
    sentiment_score=sentiment.polarity_scores(sentence) #inbuild function to analyze  
    
    #positive marks
    pos= str(sentiment_score['pos']*100 ) + '% Positive'
    positive_mark.insert(10,pos) 
    
    #neutral marks
    neu= str(sentiment_score['neu']*100 ) + '% Neutral'
    neutral_mark.insert(10,neu)
    
    #negative marks
    neg= str(sentiment_score['neg']*100 ) + '% Negative'
    negative_mark.insert(10,neg)
 
    #overall mark, compound is aggregated score between -1 to 1
    if sentiment_score['compound'] < -0.05:
        overall='Negative'

    elif sentiment_score['compound'] > 0.05:
        overall='Positive'
        
    else:
        overall='Neutral'
        
    overall_mark.insert(10, overall)

    
def clearall():
    
    #delete the contents from entry widget
    positive_mark.delete(0, END)
    negative_mark.delete(0, END)
    neutral_mark.delete(0,END)
    overall_mark.delete(0,END)
    data.delete(1.0, END)
    
#Create labels, buttons, text
enter_sent=Label(root, text='Enter your Sentence', font=('vardana', 12, 'bold'), bg='light green')
enter_sent.grid(row=0, column=2, columnspan=2, padx=20, pady=10)

data=Text(root, width=65, height=5, borderwidth=2 )
data.grid(row=2, column=2, columnspan=2,padx=40, pady=5)

check=Button(root, text='Check Sentiment', font=('vardana', 12, 'bold'), bg='purple', fg='white',command=detect_sentiment)
check.grid(row=4, column=2, columnspan=2, padx=60, pady=5)

positive=Label(root, text='Sentence rated for Positive as:', font=('vardana', 12), bg='light green')
positive.grid(row=6, column=2, columnspan=2, padx=20, pady=10)

positive_mark=Entry(root, width=30, borderwidth=2)
positive_mark.grid(row=8, column=2, columnspan=2, padx=50)

neutral=Label(root, text='Sentence rated for Neutral as:', font=('vardana', 12), bg='light green')
neutral.grid(row=10, column=2, columnspan=2, padx=20, pady=10)

neutral_mark=Entry(root, width=30, borderwidth=2)
neutral_mark.grid(row=12, column=2, columnspan=2, padx=50)

negative=Label(root, text='Sentence rated for Negative as:', font=('vardana', 12), bg='light green')
negative.grid(row=14, column=2, columnspan=2, padx=20, pady=10)

negative_mark=Entry(root, width=30, borderwidth=2)
negative_mark.grid(row=16, column=2, columnspan=2, padx=50)

overall=Label(root, text='Sentence rated Overall as:', font=('vardana', 12), bg='light green')
overall.grid(row=18, column=2, columnspan=2, padx=20, pady=10)

overall_mark=Entry(root, width=30, borderwidth=2)
overall_mark.grid(row=20, column=2, columnspan=2, padx=50)

exit=Button(root, text='   Exit   ', font=('vardana', 12, 'bold'), bg='red',command=root.destroy)
exit.grid(row=20, column=1,columnspan=2, padx=50)

clear=Button(root, text='  Clear  ', font=('vardana', 12, 'bold'), bg='blue', command= clearall)
clear.grid(row=20, column=3,columnspan=2, padx=50)



    
    
root.mainloop()
```
